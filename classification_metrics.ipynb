В контексте бинарной классификации существует несколько важных метрик, позволяющих провести оценку результатов.

**Confusion matrix** позволяет получить первую основную оценку смоделированных результатов:

Рассмотрим наиболее часто используемые метрики оценки бинарной классификации.

## <font color='mediumblue'>**Accuracy**</font>

---


Позволяет сделать вывод об общей производительности модели. Чувствительна к дисбалансу классов.

\begin{align}
\mathbf{Accuracy} = \frac{\mathbf{TP} + \mathbf{TN}}{\mathbf{TP} + \mathbf{TN} + \mathbf{FP} + \mathbf{FN}}
\end{align}

## <font color='mediumblue'>**Precision**</font>

---


Позволяет сделать вывод о том, насколько точны позитивные прогнозы.

\begin{align}
\mathbf{Precision} = \frac{\mathbf{TP}}{\mathbf{TP} + \mathbf{FP}}
\end{align}

## <font color='mediumblue'>**Recall (Sensitivity)**</font>

---


Позволяет оценить охват фактической **положительной** выборки в прогнозах модели.

\begin{align}
\mathbf{Recall} = \frac{\mathbf{TP}}{\mathbf{TP} + \mathbf{FN}}
\end{align}

## <font color='mediumblue'>**Specificity**</font>

---


Позволяет оценить охват фактической **негативной** выборки в прогнозах модели.

\begin{align}
\mathbf{Specificity} = \frac{\mathbf{TN}}{\mathbf{TN} + \mathbf{FP}}
\end{align}

## <font color='mediumblue'>**F1 score**</font>

---


Гибридная метрика, применяемая для оценки качества результатов моделирования на несбалансированных классах.

\begin{align}
\mathbf{F1 \, score} = \frac{\mathbf{2} \cdot \mathbf{TP}}{\mathbf{2} \cdot \mathbf{TP} + \mathbf{FP} + \mathbf{FN}}
\end{align}

## <font color='mediumblue'>**ROC**</font>

---


ROC (receiver operating curve) - график зависимости **True Positive Rate** от **False Positive Rate** при изменениях пороговых значений.

ROC-кривая графически изображает контраст между показателями истинно положительных и ложноположительных результатов при различных пороговых значениях.

У хорошего классификатора кривая лежит по большей части либо целиком выше прямой y=x. Это связано с тем что при хорошей классификации получают максимальный TPR при минимальном FPR.

True Positive Rate является **эквивалентом** метрики Recall (Sensitivity):

\begin{align}
\mathbf{True \, Positive \, Rate \, (TPR)} = \frac{\mathbf{TP}}{\mathbf{TP} + \mathbf{FN}}
\end{align}

False Positive Rate является **эквивалентом** 1 - Specificity

\begin{align}
\mathbf{False \, Positive \, Rate \, (FPR)} = \frac{\mathbf{FP}}{\mathbf{TN} + \mathbf{FP}}
\end{align}

## <font color='mediumblue'>**AUC**</font>

---


Показатель AUC (Area Under the ROC Curve) — это мера, которая позволяет суммировать производительность модели одним числом, измеряя площадь под кривой ROC. AUC колеблется от 0 до 1, где более высокое значение AUC указывает на более высокую производительность модели.

Значение 0,5 демонстрирует непригодность выбранного метода классификации (соответствует случайному гаданию). Значение менее 0,5 говорит, что классификатор действует с точностью до наоборот: если положительные назвать отрицательными и наоборот, классификатор будет работать лучше.
